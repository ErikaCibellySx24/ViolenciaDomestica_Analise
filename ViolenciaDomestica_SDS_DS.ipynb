```python
# realizando as importações das bibliotecas utilizadas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# realizando leitura do arquivo
df =  pd.read_excel("/content/MICRODADOS_DE_VIOLÊNCIA_DOMÉSTICA_JAN_2015_A_SET_2023.xlsx");
```

```python
# realizando filtragem pelo sexo feminino
df_fem = df.loc[df['SEXO'].str.contains('FEMININO')]
```

```python
# verificando as informações do novo dataframe
df_fem.info()
```

```python
df_fem.head(1-2)
```

```python
# renomeando colunas
df_fem = df_fem.rename(columns={"MUNICÍPIO DO FATO": 'MUNICÍPIO',"REGIAO GEOGRÁFICA": 'REGIAO',"DATA DO FATO" : "DATA OCORRENCIA","IDADE SENASP" : 'IDADE'} )
df_fem.columns = df_fem.columns.str.upper()
```

```python
# renomeando idades
df_fem['IDADE'] = df_fem['IDADE'].replace({"3) 18-24": "Entre 18 e 24 anos", "6) 35-64": "Entre 35 e 64 anos", "7) 65 OU MAIS": "65 ou mais", "4) 25-29": "Entre 25 e 29 anos", "1) 00-11": "Entre 0 e 11 anos", "2) 12-17": "Entre 12 e 17 anos", "5) 30-34": "Entre 30 e 34 anos"})
```

```python
df_fem.head(1-2)
```

```python
# verificando quantidade de números nulos
df_fem.isnull().sum()
```

```python
# verificando quais são as regiões com maior indíce de idades nulas
dados_idade_nula = df_fem[df_fem['IDADE'].isnull()]

contagem_por_regiao = dados_idade_nula['REGIAO'].value_counts()

contagem_por_regiao.plot(kind='bar')
plt.title('Número de Idades Nulas por Região')
plt.show()
```

```python
# realizando soma de envolvidos nos casos de violência doméstica
soma_envolvidos = df_fem['TOTAL DE ENVOLVIDOS'].sum()
print("A soma total de envolvidos em violência domésticas onde as vítimas eram do sexo feminino foi de", soma_envolvidos)
```

```python
# removendo última parte do texto que se repetia
df_fem['NATUREZA'] = df_fem['NATUREZA'].str.replace('POR VIOLÊNCIA DOMÉSTICA/FAMILIAR', '')
```

```python
dados_idade_nula = df_fem[df_fem['IDADE'].isnull()]

contagem_por_natureza = dados_idade_nula['NATUREZA'].value_counts()

contagem_por_natureza.plot(kind='bar')
plt.title('Número de Idades Nulas por natureza')
plt.show()
```

```python
df_fem['DATA OCORRENCIA'] = pd.to_datetime(df_fem['DATA OCORRENCIA'])

df_fem['ANO'] = df_fem['DATA OCORRENCIA'].dt.year

contagem_por_ano = df_fem['ANO'].value_counts().sort_index()

contagem_por_ano.plot(kind='line', marker='o', linestyle='-', color='b')
plt.title('Número de ocorrências por ano')
plt.show()
```

### Conclusões

Entre tantas pesquisas e tantas leituras, evidenciamos que um dos motivos de mulheres não denunciarem seus agressores são esses listados abaixo:

1. Medo de retaliação: As vítimas podem temer retaliação por parte do agressor se denunciarem o abuso. Isso pode incluir ameaças de violência física adicional ou até mesmo de prejudicar outras pessoas próximas à vítima.

2. Dependência financeira: Algumas mulheres podem depender financeiramente do agressor e podem temer perder suporte financeiro se denunciarem o abuso. Isso pode dificultar sua capacidade de sair do relacionamento abusivo.

3. Vergonha e estigma: O estigma social associado à violência doméstica pode fazer com que as vítimas sintam vergonha e evitem falar sobre sua situação. Elas podem temer o julgamento dos outros e, por isso, escolhem sofrer em silêncio.

Visto esses fatores que criam barreiras para que essas mulheres se sintam seguras e amparadas para realizar a denuncia. Ainda há muitas mulheres que  denunciam.

Essas vitimas podem escolher diferentes métodos para denunciar a violência doméstica, seja fazendo uma denúncia anônima, procurando ajuda de amigos, familiares, vizinhos, ou buscando apoio de organizações e profissionais especializados.


Para entender melhor o porque disso ocorre, fizemos uma busca e encontramos algumas reportagens que fizeram alguns estudos de caso.

Alguns fatores evidenciados são:

1. Fatores sazonais: Alguns estudos sugerem que essas flutuações podem estar relacionadas a fatores como o clima, o aumento de eventos sociais, festas e comemorações durante os primeiros meses do ano, que podem contribuir para um ambiente propício à violência.

2. Tensões familiares: O início do ano pode ser um período de maior tensão para algumas famílias devido a fatores como pressões financeiras após as festas de final de ano, expectativas elevadas e conflitos interpessoais. Essas tensões podem aumentar a incidência de violência doméstica.

3. Denúncias: Pode ser que as denúncias de violência contra a mulher aumentem nos primeiros meses do ano, devido a campanhas de conscientização ou a fatores específicos, como mulheres que buscam ajuda após as festas de final de ano. Esse aumento nas denúncias pode dar a impressão de que os crimes estão ocorrendo com mais frequência, embora eles já estivessem ocorrendo anteriormente.

4. Condições socioeconômicas: Fatores econômicos e sociais, como desemprego, pobreza e instabilidade financeira, podem influenciar a ocorrência de crimes, incluindo a violência contra a mulher. Os primeiros meses do ano podem ser um momento de maior vulnerabilidade para algumas famílias devido a esses fatores.
